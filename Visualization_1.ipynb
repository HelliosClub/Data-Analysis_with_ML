```python
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
data = pd.read_excel('04_data1.xlsx', index_col = 'area')
data.head()
```

```python
data2020_good = data['2020_good']
```

```python
plt.figure(figsize=(20,4)) ##새로운 그림 만들고 가로 세로 인치를 정한다.
plt.plot(data2020_good, linestyle = '--', color = 'b', marker='o') ##
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2020 Fine Dust(pm10) Good Line Graph')
plt.grid()
plt.show()
```

```python
print(data.columns)
```

```python
plt.figure(figsize=(20,4))

for year in range(2015,2021):
  year = str(year)+'_good'
  chartdata = data[year]
  plt.plot(chartdata, marker = 'o', label = year)

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015~2020 Fine Dust(pm10) Good Line Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,4))
import numpy as np
index = np.arange(7)

data6 = data.loc['Seoul':'Ulsan','2015_good':'2020_good']
for year in range(2015,2021):
  year = str(year)+'_good'
  chartdata = data6[year]
  plt.bar(index, chartdata, width=0.15, label= year)
  index = index +0.15

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.xticks(index-0.53,['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2015~2020 Fine Dust(pm10) Good Group Bar Graph')
plt.legend()
plt.show()
```

```python
data2 = pd.read_csv('04_data2.csv')
```

```python
##산점도 그래프 그리기

PetalLength = data2['PetalLength']
PetalWidth = data2['PetalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Petal Scatter Graph')
plt.grid()
plt.show()
```

```python
PetalLength = data2['PetalLength']
PetalWidth = data2['PetalWidth']
SepalLenth = data2['SepalLength']
SepalWidth = data2['SepalWidth']

plt.figure(figsize=(15,4))
plt.plot(PetalLength, color = 'b', label = 'Petal Length',marker = 'o')
plt.plot(PetalWidth, color = 'r', label = 'Petal Width', marker = 'o')
plt.plot(SepalLenth, color = 'g', label = 'Sepal Length',marker = 'o')
plt.plot(SepalWidth, color = 'y', label = 'Sepal Width', marker = 'o')

plt.xlabel('Iris')
plt.ylabel('Length')
plt.title('Iris Line Graph')
plt.grid()
plt.show()
```
