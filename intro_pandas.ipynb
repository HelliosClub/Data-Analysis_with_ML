```python
import pandas as pd
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips[tips['tip']>= 7.00]
```

```python
conditions = (tips['sex']== 'Female')& (tips['smoker'] == 'Yes')
tips[conditions].head()
```

```python
tips[tips['tip'] == tips['tip'].max()]
```

```python
tips.sort_values(['total_bill','tip'],ascending = False).head(10)
```

```python
tips.sort_values(['size','tip'],ascending=[False, True]).head(10)
```

```python
tips['total']= tips['total_bill']+tips['tip']
tips
```

```python

tips.groupby('day')['tip'].mean()

```

```python
tips[(tips['day'] == 'Fri') |( tips['day'] == 'Sat')]
```

```python
tips[tips['tip']> tips['tip'].mean()]
```

```python
tips.loc[:,['total_bill','tip','time','total']]
```
