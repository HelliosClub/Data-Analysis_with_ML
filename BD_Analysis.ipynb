```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

drama = pd.read_csv('/content/top100_kdrama.csv')
drama.head()
```

```python
drama.info()
```

```python
drama[['Name', 'Rating']]
```

```python
d = drama['Number of Episode'] == drama['Number of Episode'].max()
drama[d]
```

```python
d = drama['Network'].str.contains('Netflix')
drama[d].head()
```

```python
d = drama.groupby('Year of release')['Rating'].mean()
pd.DataFrame(d)
```

```python
top5=drama['Year of release'].value_counts().head(5)
plt.figure(figsize=(16,10))
plt.bar(top5.index,top5,color='g')
plt.xlabel('Year of release')
plt.ylabel('Genre')
plt.grid()
plt.show()
```
